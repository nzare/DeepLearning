# Simple Linear Regression. Minimal example

### Import the relevant libraries

```python
# We must always import the relevant libraries for our problem at hand. NumPy is a must for this example.
import numpy as np

# matplotlib and mpl_toolkits are not necessary. We employ them for the sole purpose of visualizing the results.  
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Generate random input data to train on

```python
observations=1000

xs=np.random.uniform(-10,10,size=(observations,1))   # 1000x1
zs=np.random.uniform(-10,10,size=(observations,1))   # 1000x1

inputs=np.column_stack((xs,zs))  # 1000x2
print(inputs.shape)
```

### Create targets

```python
noise=np.random.uniform(-1,1,size=(observations,1)) #1000x1

targets= 2*xs - 3* zs +5 + noise  #1000x1

print(targets.shape)
```

### Plot the training data

```python
targets= targets.reshape(observations,)
fig=plt.figure()
ax= fig.add_subplot(111,projection='3d')
ax.plot(xs,zs,targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('targets')
ax.view_init(azim=100)
plt.show()
targets=targets.reshape(observations,1)
```

### Initialize varibles

```python
init_range=0.1

weights=np.random.uniform(-init_range,init_range,size=(2,1))  #2x1
bias=np.random.uniform(-init_range,init_range,size=(1,1))  #1x1
print(bias)
print(weights)
```

### Define learning rate

```python
learning_rate=0.02
```

### Train the model

```python
for i in range(100):
    outputs= np.dot(inputs,weights) + bias  #1000x1
    deltas= outputs-targets   #1000x1
    loss = np.sum(deltas**2)/2/observations
    print(loss)
    
    deltas_scaled=deltas/observations
    
    weights=weights-learning_rate*np.dot(inputs.T,deltas_scaled)
    bias=bias-learning_rate*np.sum(deltas_scaled)
    
    
```

### Print weights and the biases

```python
print(weights,bias)
```

### Plot last weights vs targets

```python
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
